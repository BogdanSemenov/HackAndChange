```python
import numpy as np

def replace_binary_value(b):
    if b and pd.notnull(b):
        return 1
    else:
        return -1
```

```python
ds = ds[ds["CITY"] != "КАЛУГА"]
ds.describe()
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split 
from scipy.optimize import minimize
from sklearn.utils import shuffle

data_not_kaluga = ds[ds["CITY"] != "КАЛУГА"]
X = data_not_kaluga[data_not_kaluga['cashbox_daily'].notnull()].loc[:, ['cashbox_MAX']]
X = (X - X.mean()) / X.std() # mean = 0; std = 1
y = data_not_kaluga[data_not_kaluga['cashbox_daily'].notnull()]['cashbox_daily']
y = (y - y.mean()) / y.std()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
target_names = ['Postamat_daily', 'cashbox_daily']
features_names = ['Postamat_trend', 'cashbox_MAX', 'cashbox_trend', 'hh_500', 'value1', 'value2', 'POPULATION', 'NEAR_Malls', 'NEAR_Business_centers', 'NEAR_metro_rjd', 'NEAR_Stations', 'macro_salary_avg_yearly']
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
plt.scatter(X['cashbox_MAX'], y)
plt.scatter(X_test['cashbox_MAX'], lin_reg.predict(X_test))
```
